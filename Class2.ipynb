```python
class Stack():

  def __init__(self):
    self.top = None


  class Node():
    def __init__(self, element) -> None:
      self.data = element
      self.bellow = None

    def __str__(self) -> str:
      return str(self.data)

  def push(self, element):
    new_node = self.Node(element)
    if not self.top:
      self.top = new_node
    else:
      old_top = self.top
      new_node.bellow = old_top
      self.top = new_node


  def pop(self):
    if self.top:
      old_top = self.top
      self.top = old_top.bellow
      return old_top.data
    raise IndexError("Empty stack")

  def size(self):
    return self.__length


  def is_empty(self):
    if self.top is None:
      return True
    return False

  def peek(self):
    if not self.is_empty():
      return self.top.data
    return None
```

```python
class Queue():
  def __init__(self) -> None:
    self.items = []

  def enqueue(self, element):
    self.items.insert(0, element)

  def dequeue(self):
    return self.items.pop()

  def size(self):
    return len(self.items)

  def is_empty(self):
    return self.items == []

  def peek(self):
    return self.items[len(self.items-1)]
```

## Implement QueueII, a "from scratch" implementation of Queue.

### Criteria
Your Queue class should, at a minimum cover the following features:
1. Must be able to enqueue (from one end)
2. Must be able to dequeue (from the end opposite to where elements are enqueued)
3. Must have a size method that returns how many elements are in the queue
4. Must have an is_empty method that returns `True` if the queue is empty, `False` otherwise
5. Must have a peek method that returns the topmost `value` in the queue

## Note
You should not use any built-in data structures, relying solely on Nodes or any other custom built data structures and methods.
