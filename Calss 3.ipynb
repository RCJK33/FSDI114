```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        # TODO :
        self.tail = None

    class __Nodo:
        def __init__(self, data):
            self.data = data
            self.next = None

        def __str__(self) -> str:
            return str(self.data)

    def append(self, value):
        new_nodo = self.__Nodo(value)
        if self.head is None:
            self.head = new_nodo
            self.tail = self.head
        else:
            self.tail.next = new_nodo
            self.tail = new_nodo


    def insert(self, index, value):        
        new_nodo = self.__Nodo(value)
        if self.head is None:
            self.head = new_nodo
            self.tail = self.head
            return
        
        if index == 0:
            new_nodo.next = self.head
            self.head = new_nodo
            return

        nodo = self.head
        for i in range(0,index-1):
            if nodo.next is None:
                break
            nodo = nodo.next

        new_nodo.next = nodo.next
        nodo.next = new_nodo


    def remove(self, value):
        #remove the first occurrence of value, ValueError if not present
        found = False
        if self.head:
            current = self.head
            prev = None
            while current.next and not found:
                if current.data == value:
                    found = True
                else:
                    prev = current
                    current = current.next
            if prev and found:
                prev.next = current.next
            if not prev and found:
                self.head = None
        raise ValueError("Target value not in list")

    def index(self, value):
        current = self.head
        index = 0
        found = False
        while not found:
            if current.data == value:
                return index
            index += 1
            current = current.next

    def access(self, index):
        current = self.head
        for i in range(index):
            current = current.next
        return current.data

    def __str__(self) -> str:
        out = "["
        
        if self.head is None:
            return "[]"

        nodo = self.head
        while nodo:
            out += str(nodo)
            if not nodo.next is None:
                out += ", "
            nodo = nodo.next
        out += "]"

        return out
```

```python
sll = SinglyLinkedList()
for i in "Hello, world":
    sll.append(i)

sll.insert(1, "!")

print(sll)
```
